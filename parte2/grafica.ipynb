```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv('Reporte.csv',sep=" ", header=None).astype("int")
df = datos.to_numpy()
```

```python
n = (int)(datos.shape[0]/4)
mats = np.split(df,n,axis=0)
mats
```

```python
n = df.shape[0]/(4*2)
pos = 0
columnas = ['Ants', 'MotorBikes', 'SunFlower','Watches']
for i in range(0,int(n)):
    cmap = 'Paired'
    #cmap = 'Blues'
    #cmap = 'BuPu'
    fig, axes = plt.subplots(1, 2,figsize=(15,6))
    title='Prueba '+str(i+1)
    fig.suptitle(title)
    axes[0].set_title('Predicción GrayScale')
    axes[1].set_title('Predicción BGR')

    mat_gray = mats[pos]
    df_cm_gray = pd.DataFrame(mat_gray , index=columnas, columns=columnas)
    grafica_gray=sns.heatmap(df_cm_gray, cmap=cmap, annot=True,ax=axes[0])
    
    mat_color = mats[pos+1]
    df_cm_color = pd.DataFrame(mat_color , index=columnas, columns=columnas)
    grafica_color=sns.heatmap(df_cm_color, cmap=cmap, annot=True,ax=axes[1])

    plt.ylabel('Test')
    plt.xlabel('Train')
    grafica_gray.set(xlabel='Train', ylabel='Test')
    grafica_color.set(xlabel='Train', ylabel='Test')
    plt.savefig(title+".png")
    plt.show()
    pos+=2
```
